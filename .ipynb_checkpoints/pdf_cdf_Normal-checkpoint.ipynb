```python
import numpy as np
import seaborn as sns

# Generate 1000 samples with 60 as its mean and 10 as its std
a = np.random.normal(60, 10, 1000)
sns.distplot(a, hist=True, kde=True, bins=20)

print(len(a))
print(np.mean(a))
print(np.std(a))
print(np.var(a))
```

```python
def gt_70():
    count = 0
    for num in a:
        if num >= 70:
            count += 1
    return count/len(a)

print(gt_70())
```

```python
count = 0
def add(num):
    return num + 1
gt = sum([add(count) for num in a if num >=70]) / len(a)
print(gt)
```

```python
from scipy.stats import norm

1 - norm.cdf(70, loc=60, scale=10)

# or

print(norm.sf(70, loc=60, scale=10))
```

```python
x = np.random.normal(60,10,100000)
sns.distplot(x,hist=True, kde=True, bins=40)

y = np.random.normal(60,20,100000)
sns.distplot(y,hist=True, kde=True, bins=40)
```

```python
norm.cdf(70, loc=60, scale=10) - norm.cdf(50, loc=60, scale=10)
```

```python
def between_val(ls, low, high):
    count = 0
    for num in ls:
        if num >= low and num <= high:
            count += 1
    return count / len(ls)

def percent_between(ls, low, high):
    return len([i for i in ls if low <= i and i<= high])/len(ls)

print(between_val(x, 50, 70))
print(between_val(y, 40, 80))
print()
print(percent_between(x, 50, 70))
print(percent_between(y, 40, 80))
```
